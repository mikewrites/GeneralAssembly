```python
sat_2017['rounded'] = sat_2017['rounded'].apply(lambda x : int(x))
sat_2019['rounded'] = sat_2019['rounded'].apply(lambda x : int(x))
sat_2018['rounded'] = sat_2018['rounded'].apply(lambda x : int(x))
```

```python
sat_percentiles = pd.read_pickle('../data/sat_percentiles.pkl')
```

```python
def add_percentiles_sat(df, year):
    df.reset_index(drop = True, inplace = True)
    df['percentile'] = 0
    for i in range(len(df)):
        rounded = df.loc[i, 'rounded']
        df.loc[i, 'percentile'] = sat_percentiles.loc[rounded, f'{year}']
    return df
```

```python
sat2018 = add_percentiles_sat(sat_2018, 2018)
```
