Lets look at some of the ordinal and categorical values with groupby:



```python
def plot_val(col_name, df=ames):
    to_plot = df.loc[:, (col_name, 'SalePrice')]
    to_plot[col_name].fillna('None', inplace = True)
    to_plot.groupby(col_name).mean('SalePrice')
    sns.barplot(x= col_name, y= 'SalePrice', data = to_plot)
    plt.show()
    display(to_plot.value_counts(col_name))
```

```python
def plot_val_big(col_name, df=ames):
    to_plot = df.loc[:, (col_name, 'SalePrice')]
    to_plot[col_name].fillna('None', inplace = True)
    to_plot.groupby(col_name).mean('SalePrice')
    to_plot = to_plot.sort_values('SalePrice', ascending = True)
    sns.set(rc={"figure.figsize":(10, 15)})
    sns.barplot(y= col_name, x= 'SalePrice',data = to_plot)
    p25 = np.percentile(ames['SalePrice'], 25)
    p75 = np.percentile(ames['SalePrice'], 75)
    plt.axvline(ames['SalePrice'].mean())
    plt.axvline(p25)
    plt.axvline(p75)
    plt.show()
    display(to_plot.value_counts(col_name))
```

# MS Zoning

based on the documentation. all values with R are zoned residential while FV is zoned "floting villiage residential". Looking at the graph, it appears that residential zoning and non-residential are both pretty simmilar means, but quite different from eachother. I'm going to double check that explicitly

```python
def plot_val_big_zng(col_name, df=ames):
    to_plot = df.loc[:, (col_name, 'SalePrice')]
    to_plot[col_name].fillna('None', inplace = True)
    to_plot.groupby(col_name).mean('SalePrice')
    to_plot = to_plot.sort_values('SalePrice', ascending = True)
    sns.set(rc={"figure.figsize":(10, 15)})
    sns.barplot(y= col_name, x= 'SalePrice',data = to_plot)
    p25 = np.percentile(ames['SalePrice'], 25)
    p75 = np.percentile(ames['SalePrice'], 75)
    plt.axvline(ames['SalePrice'].mean())
    plt.axvline(p25)
    plt.axvline(p75)
    plt.show()
    display(to_plot.value_counts(col_name))
```

```python
plot_val_big_zng('MS Zoning')
```

Yup, no non-residentially zoned property hits even the 25%ile I'll convert this to residential and non residential data, in my feature engineering notebook
